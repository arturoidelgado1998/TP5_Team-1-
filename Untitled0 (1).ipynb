#Introduction

The following analyses cross-examine several variables determined from our survey data, including: the perceived quality of Disney+ comparative to consumers' streaming service preferences, the role of gender comparative to consumers' streaming service preferences, and the level of recommendation consumers would provide relative to the perceived quality of Disney+. Corresponding to each cross-tab are the variables' respective levels of association, complete with Chi-square and Cramer's V tests, culminating in each cross-examinations' p-value. From this data, we hope to determine more concrete associations among differing variables in an effort to pinpoint how Disney+ can more effectively target college students and young adults, with the ultimate goal to become the leading streaming service within the crowded VOD industry.



```python
!pip install researchpy
import numpy as np
import pandas as pd
import researchpy as rp
from glob import glob
from scipy import stats as st

def concat (outfile = 'DisneyCSVParsed2.csv'):
  filenames = glob('./sample_data/DisneyCSVParsedFinal10.csv')
  dataframes = []
  for files in filenames:
    dataframes.append(pd.read_csv(files))
  concatDF = pd.concat(dataframes, axis = 0)
  concatDF.to_csv('./sample_data/'+ outfile, index = None)

datasetoriginal =  pd.read_csv("sample_data/DisneyCSVParsedFinal10.csv")
display(datasetoriginal.head(6))
def Male(row):
  if "Male" in row.Gender:
    return 1
  else:
    return 0
def Female(row):
  if "Female" in row.Gender:
    return 1
  else:
    return 0

def ServicesToMe(row):
  if "Moderately important" in row.StreamingServices:
    return 3
  elif "Not at all important" in row.StreamingServices:
    return 1
  elif "Very important" in row.StreamingServices:
    return 4
  elif "Extremely important" in row.StreamingServices:
    return 5
  else:
    return 2
datasetoriginal['Male'] = datasetoriginal.apply(Male,axis=1)
datasetoriginal['Female'] = datasetoriginal.apply(Female,axis=1)
datasetoriginal['ServicesToMe'] = datasetoriginal.apply(ServicesToMe, axis=1 )
display(datasetoriginal)
```

#Cross-Tabs 1) Quality of Disney+ Vs. Consumer's Best Sreaming Prefernces

```python
ind_obs2 , result3 = rp.crosstab(datasetoriginal["The quality of the Disney+ website/app is..."], datasetoriginal["Which streaming service do you believe is the best? - Selected Choice"], test = 'chi-square')
display(ind_obs2)
display(result3)
```

#Cross-Tabs 1) Low to Moderate Association
- 10/50(20%) college consumers find that Disney+ is both their favorite and is excellent/good - 20% is pretty good in relation to all other streaming services
- 21/50(62%) college consumers find that Disney+ is excellent/good, in addition, they find that HBO Max, Hulu, and Netflix are on the same level of excellence - This shows that HBO Max, Hulu, and Netflix are competitiors Disney+ needs to look out for when trying to target college consumers.
